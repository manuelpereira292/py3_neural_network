```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_p(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
X = np.linspace(-5, 5, 10)
Y = sigmoid(X)
```

```python
Y
```

```python
plt.plot(X, sigmoid(X), c='b')
```

```python
X = np.linspace(-6, 6, 100)
plt.plot(X, sigmoid(X), c='r')
plt.plot(X, sigmoid_p(X), c='b')
```
